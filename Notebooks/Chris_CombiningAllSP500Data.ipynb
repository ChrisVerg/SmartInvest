```python
import pandas as pd
```

```python
set1 = pd.read_csv('../Data/all_3_Ig.csv')
```

```python
set2 = pd.read_csv('../Data/all_4_JV.csv')
```

```python
set3 = pd.read_csv('../Data/all_4.JC.csv')
```

```python
SP500total = pd.merge(set1, pd.merge(set2, set3, on='year_month'), on='year_month')
```

```python
SP500total.head()
```

```python

columns_to_drop = ['Unnamed: 0_x', 'Unnamed: 0_y', 'index']
SP500 = SP500total.drop(columns=columns_to_drop)
```

```python
SP500.columns
```

```python
# 
column_names = {
    'diff_sp_mat': 'Materials',
    'diff_sp_utilities': 'Utilities',
    'diff_sp_real_estate': 'Real Estate',
    'diff_financials': 'Financials',
    'diff_IT': 'Information Technology',
    'diff_Healthcare': 'Healthcare',
    'diff_CD': 'Consumer Discretionary',
    'diff_energy': 'Energy',
    'diff_Industrials': 'Industrials',
    'diff_CommS': 'Communication Services',
    'diff_CStaples': 'Consumer Staples'
}

SP500 = SP500.rename(columns=column_names)
```

```python
SP500.head()
```

```python
SP500.to_csv('../Data/SP500.csv')
```
