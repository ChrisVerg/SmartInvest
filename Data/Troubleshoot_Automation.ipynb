```python
# Automate Collection

# This block of code improves the format of the data collection

# NLP Remove stop words
import nltk
from nltk.corpus import stopwords

import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup

# Add date and time to file
import datetime

```

```python
def extract_headlines(url):
    global df
    
    # Fetch the data from the URL
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    # Extract the headlines
    headlines = set()
    for headline in soup.find_all('div', class_="FeaturedCard-contentText"):
        links = headline.find_all("a")
        for link in links:
            headlines.add(link.text.strip())

    for headline in soup.find_all('div', class_="SecondaryCard-headline"):
        links = headline.find_all("a")
        for link in links:
            headlines.add(link.text.strip())

    for headline in soup.find_all('div', class_="LatestNews-headlineWrapper"):
        links = headline.find_all("a")
        for link in links:
            headlines.add(link.text.strip())

    # Create a DataFrame object with the headlines
    df = pd.DataFrame(list(headlines), columns=['Headlines'])
    
    return df

# Example usage:
url = "https://www.cnbc.com/"
collected_headlines = extract_headlines(url)
```

```python
##                                                This Seems to be the winner

import pandas as pd
import datetime

def add_new_values_to_dataframe(file_path):
    global df  # Access the global df variable

    # Read the existing data from the CSV file
    existing_data = pd.read_csv(file_path, index_col=0)

    # Check if there are existing columns
    if len(existing_data.columns) > 0:
        latest_column = existing_data.columns[-1]
        existing_values = set(existing_data[latest_column].dropna())

        # Get new values that are not already in the DataFrame
        new_values = [value for value in df['Headlines'] if value not in existing_values]

        if len(new_values) > 0:
            # Add new values in a new column
            current_datetime = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            # Ensure the dimensions match by extending or truncating the new values list
            if len(new_values) < len(existing_data.index):
                new_values.extend([None] * (len(existing_data.index) - len(new_values)))
            elif len(new_values) > len(existing_data.index):
                new_values = new_values[:len(existing_data.index)]

            existing_data[current_datetime] = new_values

    else:
        # If no existing columns, add all values in a new column
        current_datetime = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        existing_data[current_datetime] = df['Headlines']

    # Save the updated data to the CSV file
    existing_data.to_csv(file_path)

# Example usage:
file_path = 'C:/Users/chris/Desktop/GitHub/SmartInvest/Data/TestFormat_CV.csv'
add_new_values_to_dataframe(file_path)
```
